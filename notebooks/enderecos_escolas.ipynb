### Get adresses

O endereço e CEP da escola podem ser obtidos através do [Localização escolar - IDEB](http://idebescola.inep.gov.br/ideb/consulta-publica). 

Cada escola tem uma página com a URL: `http://idebescola.inep.gov.br/ideb/escola/dadosEscola/[cod_escola]`. 

Somente as escolas que têm Ideb ou cadastro no Censo da Educação Básica 2017 possuem esses dados no site.

```python
def get_adress(url):
    
    print(url)
    
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    
    # table of school data
    print(soup.find_all('table'))
    s = soup.find_all('table')[0].get_text()
    
    rua = str(re.search('EndereÃ§o(.*)Bairro', s).group(1)).strip()
    bairro = str(re.search('Bairro(.*)CEP', s).group(1)).strip()
    cep = str(re.search('CEP(.*)MunicÃ\xadpio', s).group(1)).strip()
    
    return rua, bairro, cep

#ideb['rua'], ideb['bairro'], ideb['cep'] = ideb['cod_escola_url'].apply(get_adress)
```
